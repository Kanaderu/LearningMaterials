```python
from sklearn.datasets import fetch_20newsgroups
```

```python
classes = ['alt.atheism']
corpus = fetch_20newsgroups(subset='train', categories=classes, shuffle=True, random_state=42)
```

```python
print(len(corpus.filenames))
print(len(corpus.data))
print(f'number of labels: {len(corpus.target_names)}')
print(f'list of labels: {corpus.target_names}')
corpus.data[0:5]
```

```python
for d in range(0,6):
    print(corpus.data[d])
    print(corpus.target_names[corpus.target[d]])
    print(corpus.target[d])
```
