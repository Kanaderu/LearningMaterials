```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
from sklearn import metrics

classes = ['alt.atheism', 'soc.religion.christian',
            'comp.graphics', 'sci.med']

corpus = fetch_20newsgroups(subset='train', categories=classes,
                            shuffle=True, random_state=42)

for d in range(6):
    print(corpus.data[d])
    print('Label: ', corpus.target_names[corpus.target[d]], 'target: ', corpus.target[d])

vectorizer1 = CountVectorizer(analyzer='word', ngram_range=(1,1),
                              stop_words='english')
X = vectorizer1.fit_transform(corpus.data)

tfidf_transformer = TfidfTransformer()
X_tfidf = tfidf_transformer.fit_transform(X)
```

# Training classifier

```python
clf = MultinomialNB().fit(X, corpus.target)
```

```python
docs_new = ['God is love', 'OpenGL on the GPU is fast']
Y_counts = vectorizer1.transform(docs_new)
Y_tfidf = tfidf_transformer.transform(Y_counts)

predicted = clf.predict(Y_tfidf)
```

```python
for doc, category in zip(docs_new, predicted):
    print('{} => {}'.format(doc, corpus.target_names[category]))
```

# Load and run test corpus

```python
corpus_test = fetch_20newsgroups(subset='test', categories=classes, shuffle=True, random_state=42)

docs_test = corpus_test.data

test_counts = vectorizer1.transform(docs_test)
test_tfidf = tfidf_transformer.transform(test_counts)

predicted = clf.predict(test_tfidf)

print(np.mean(predicted == corpus_test.target))
print(metrics.classification_report(corpus_test.target, predicted, target_names=corpus_test.target_names))
```
